```python
import seaborn as sns
import pandas as pd
df=sns.load_dataset("tips")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
df['total_bill'].mean()
```

```python
df.groupby('sex')['total_bill'].mean()
```

```python
df.groupby('day')['total_bill'].mean()
```

```python
df.groupby('smoker')['total_bill'].mean()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
plt.title("Average Bill by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Bill")
```

```python
sns.countplot(x='day',data=df)
plt.title("Customer Count by Days")
plt.xlabel("Day")
plt.ylabel("Count")
```

```python
sns.barplot(x='smoker',y='total_bill',data=df)
plt.title("Average Bill by Smoking Status")
plt.xlabel("Smoking Status")
plt.ylabel("Total Bill")
```

```python
df[['total_bill', 'tip', 'size']].corr()
```

```python
numeric_df = df.select_dtypes(include=['number'])
corr = numeric_df.corr()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['tip_pct'] = (df['tip'] / df['total_bill']) * 100

```

```python
sns.scatterplot(x='total_bill', y='tip', hue='time', data=df)

```

**Key Insights:**


Correlation: There is a strong positive correlation (0.67) between total bill and tip, indicating that higher bills result in higher tips.

Timing: Saturday is the peak day for both spending and frequency likely due to weekend dining.

New Metric: Added tip_pct to evaluate tipping fairness across bill sizes.

Group Size:
- Larger groups increase total revenue but show a lower correlation with tip percentage.
- Total bill and group size (0.59) show a moderate relationship, meaning larger groups tend to generate higher bills.
- Tip and group size (0.49) have a moderate relationship, suggesting that group size indirectly affects tipping behavior.
- Male customers tend to spend slightly more than female customers.
- Smokers tend to spend slightly more compared to non-smokers.
